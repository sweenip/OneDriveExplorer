````python
import pandas as pd
import numpy as np
import pickle
````

````python
folder1 = 'tmp/ztsdaf1471/sweeni/241231_0107ct/'
folder2 = 'tmp/ztsdaf1471/sweeni/250101_0257ct/'
````

````python
fields = ['Type', 'fileStatus', 'diskLastAccessTime', 'diskCreationTime', 'lastChange', 'size', 'folderStatus', 'Path', 'Name', 'fileName', 'lastWriteCount']
````

````python

df1 = pd.read_csv(folder1 + '/SQLite_DB_OneDrive.csv', skipinitialspace=True, usecols=fields,
                 converters={'diskLastAccessTime': pd.to_datetime,
                             'diskCreationTime': pd.to_datetime,
                             'lastChange': pd.to_datetime
                             }, 
                 dtype={'Type' : 'category',
                        'fileStatus' : 'category',
                        'folderStatus' : 'category'}
                        )
with open(folder1 + 'out.pickle','wb') as f:
    pickle.dump(df1,f)
````

````python
df2 = pd.read_csv(folder2 + '/SQLite_DB_OneDrive.csv', skipinitialspace=True, usecols=fields,
                 converters={'diskLastAccessTime': pd.to_datetime,
                             'diskCreationTime': pd.to_datetime,
                             'lastChange': pd.to_datetime
                             }, 
                 dtype={'Type' : 'category',
                        'fileStatus' : 'category',
                        'folderStatus' : 'category'}
                        )
with open(folder2 + 'out.pickle','wb') as f:
    pickle.dump(df2,f)
````

````python
with open(folder1 + 'out.pickle','rb') as f:
    df1=pickle.load(f)
with open(folder2 + 'out.pickle','rb') as f:
    df2=pickle.load(f)
````

````python

# Id rows where Name is NaN
display(df1[(df1['Name'].isnull())])
# Delete the Scope entry row
df1.drop(df1[df1['Type'] == 'Scope'].index, inplace=True)
display(df1[(df1['Name'].isnull())])
````

````python
display(df2[(df2['Name'].isnull()) ].head())
df2.drop(df2[df2['Type'] == 'Scope'].index, inplace=True)
display(df2[(df2['Name'].isnull()) ].head())
````

````python
display(df1[(df1['Path'].isnull()) ])
display(df2[(df2['Path'].isnull()) ])
````

````python
df1['file'] = df1['Path'] + '\\' + df1['Name']
df1 = df1.drop(columns=['Path', 'Name'])
df2['file'] = df2['Path'] + '\\' + df2['Name']
df2= df2.drop(columns=['Path', 'Name'])
````

````python

# Remove this prefix from Path 	df9e9e556244467b9dabdc62d965839f
df1 = df1.replace(to_replace={'file': r'\w+\\\\'}, value='', regex=True)
df2 = df2.replace(to_replace={'file': r'\w+\\\\'}, value='', regex=True)
df1.set_index('file', drop=True, inplace=True)
df2.set_index('file', drop=True, inplace=True)
````

````python
print(df1.info())
print(df2.info())
````

````python
duplicates = df2[df2.index.duplicated()].sort_index()
duplicates.sort_values(by='lastChange', ascending=True, inplace=True)
duplicates.head()
````

````python
# Check for duplicates
duplicates = df1[df1.index.duplicated()]
display(duplicates.sort_index().head())

# print(duplicates['Type'].unique())

# Check for duplicates
duplicates = df2[df2.index.duplicated()]
display(duplicates.sort_index().head())

# print(duplicates['Type'].unique())
````

````python
result = pd.merge(df1, df2, on='file', how='outer')
# result = result.drop(columns=['folderStatus_x', 'diskCreationTime_y','Type_y','size_y','folderStatus_y'])
result = result.drop(columns=['folderStatus_x', 'folderStatus_y'])
result.info()
````

````python
result = result[['Type_x', 'Type_y',  'fileStatus_x', 'fileStatus_y', 'diskLastAccessTime_x',  'diskLastAccessTime_y',
       'lastChange_x', 'lastChange_y','fileName_x', 'fileName_y', 'lastWriteCount_x', 
       'lastWriteCount_y', 'size_x', 'size_y', 'diskCreationTime_x', 'diskCreationTime_y']]
display(result.head())
````

````python
print(result['fileStatus_x'].value_counts().index)
print(result['fileStatus_y'].value_counts().index)
````

````python
# Note that df2 has only one category '2' for fileStatus_y
# trying to compare between fileStatus_x and fileStatus_y will result in error
# TypeError: Categoricals can only be compared if 'categories' are the same

result['fileStatus_y'] = result['fileStatus_y'].cat.add_categories('3')
print(result['fileStatus_y'].cat.categories)
print(result['fileStatus_y'].value_counts().index)
````
